```python
n= int(input())
arr= [int(x) for x in input().split()]

i=0
nz=0
nt= n-1

while(i<n and i<nt):
    
    if arr[i]==0:
        arr[i] , arr[nz] = arr[nz] , arr[i]
        nz= nz+1
        i=i+1
        
    elif arr[i]==2:
        val= arr[nt]
        arr[i] , arr[nt] = arr[nt] , arr[i]
        nt= nt-1
        i=i+1
        
        if val!=1:
            i=i-1
      
    print(i, nz, nt)
```
